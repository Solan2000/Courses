```python
retaild=retail.query("Country=='Germany'")\
    .groupby('CustomerID',as_index=False)\
    .agg({'InvoiceNo':'nunique'})
qua=retaild.InvoiceNo.quantile(0.8)
retaild=retaild.query('InvoiceNo>@qua')
germany_top=retaild.CustomerID#задание 5
```

```python
top_retail_germany=retail.query('CustomerID in @germany_top')#6задание
top_retail_germany=top_retail_germany.query("StockCode!='POST'")
```

```python
top_retail_germany.groupby('StockCode',as_index=False)\
    .agg({'Quantity':'count'}).sort_values('Quantity',ascending=False)#7 задание
```
